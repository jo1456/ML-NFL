```
class generalDefense:
    def __init__(self, Fum, Ints, PenaltyYards, Pen1ds):
        self.fum = Fum
        self.ints = Ints
        self.penaltyYards = PenaltyYards
        self.pen1ds = Pen1ds

    def getList(self):
        return [self.fum,
        self.ints,
        self.penaltyYards,
        self.pen1ds]
```

```
class kicking:
	def __init__(self, FgPct, XpPct, YardsPerPunt):
		self.fgPct = FgPct
		self.xpPct = XpPct
		self.yardsPerPunt = YardsPerPunt

	def getList(self):
		return [str(self.fgPct), str(self.xpPct), str(self.yardsPerPunt)]
```

```
class kickReturns:
   def __init__(self, Punt, Kick):
       self.puntYards = Punt
       self.kickYards = Kick

   def getList(self):
        return [str(self.puntYards) , str(self.kickYards)]
   
```

```
class driveAvgs:
    def __init__(self, NumberOfDrives, Plays, ScorePercent, ToPercent, PlaysPerDrive, YardsPerDrive, Start, Time, PtsPerDrive):
        self.numberOfDrives = NumberOfDrives
        self.plays = Plays
        self.scorePercent = ScorePercent
        self.toPercent = ToPercent
        self.playsPerDrive = PlaysPerDrive
        self.yardsPerDrive = YardsPerDrive
        self.start = Start
        self.time = Time
        self.ptsPerDrive = PtsPerDrive

    def getList(self):
        return [str(self.numberOfDrives), str(self.plays), str(self.scorePercent), str(self.toPercent),
              str(self.playsPerDrive), str(self.yardsPerDrive), str(self.start)
             , str(self.time), str(self.ptsPerDrive)]
```
